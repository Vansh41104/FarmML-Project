```python
# Import libraries
from google.cloud import aiplatform
import vertexai
```

```python
# Initialize connection
aiplatform.init(location='us-central1')
```

```python
vertexai.init(project="agritwin-cv", location="us-central1")
```

```python
job = aiplatform.CustomContainerTrainingJob(
    display_name='njr_grassprediction',
    container_uri='us-central1-docker.pkg.dev/agritwin-cv/njr-grassprediction/grass_prediction@sha256:e5b8583064031fa4ad03ab5cb124addd4b26c55b9a8a583418651f78f70c791f',
    staging_bucket='njr_paddockmodel'
)
```

```python
job.run(
    replica_count=1, 
    machine_type='n1-standard-4',
    args=[
        '--input_bucket', 'njr_paddockbucket',
        '--output_bucket', 'njr_paddockbucket_output',
        '--slice_number', '1',
        '--patch_number', '2',
        '--paddock_number', '3',
        '--bq_dataset', 'njr',
        '--bq_table', 'njr_grass_predictions',
        '--model_bucket', 'njr_paddockmodel',
        '--model_prefix', 'njr_'
    ]
)
```

```python
job.run(
    replica_count=1,
    machine_type='n1-standard-4',
    accelerator_type='NVIDIA_TESLA_V100',  # or other types like 'NVIDIA_TESLA_V100', 'NVIDIA_TESLA_T4', etc.
    accelerator_count=1,  # Number of GPUs per replica
    args=['--input_bucket=njr_paddockbucket',
'--output_bucket=njr_paddockbucket_output',
'--slice_number 1',
'--patch_number 2',
'--paddock_number 3',
'--bq_dataset njr',
'--bq_table njr_grass_predictions',
'--model_bucket njr_paddockmodel',
'--model_prefix njr_']
)
```
